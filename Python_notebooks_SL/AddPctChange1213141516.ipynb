```python
import pandas as pd
from functools import reduce
```

```python
census12to17=pd.read_csv('../data_to_add_to_datast/Census2012to2017.csv')
```

```python
census12to17.head()
```

```python
census12to17.describe()
```

```python
census12to17["Year"].unique()
```

```python
census12to17_2=census12to17.replace(-66666666,0)
```

```python
census12=census12to17_2.loc[census12to17["Year"]==2012]
```

```python
census12.head()
```

```python
census13=census12to17_2.loc[census12to17["Year"]==2013]
```

```python
census13.head()
```

```python
census14=census12to17_2.loc[census12to17["Year"]==2014]
```

```python
census14.head()
```

```python
census15=census12to17_2.loc[census12to17["Year"]==2015]
```

```python
census15.head()
```

```python
census16=census12to17_2.loc[census12to17["Year"]==2016]
```

```python
census16.head()
```

```python
census1213=census12.merge(census13, left_on='Zipcode' , right_on='Zipcode', suffixes=('_12','_13'))
```

```python
census1314=census13.merge(census14, left_on='Zipcode' , right_on='Zipcode', suffixes=('_13','_14'))
```

```python
census1415=census14.merge(census15, left_on='Zipcode' , right_on='Zipcode', suffixes=('_14','_15'))
```

```python
census1516=census15.merge(census16, left_on='Zipcode' , right_on='Zipcode', suffixes=('_15','_16'))
```

```python
census1213['PerCapInc1213PctChange']=census1213['Per Capita Income_13']/census1213['Per Capita Income_12']-1
```

```python
census1213['PerCapInc1213PctChange'].head()
```

```python
census1314['PerCapInc1314PctChange']=census1314['Per Capita Income_14']/census1314['Per Capita Income_13']-1
```

```python
census1314['PerCapInc1314PctChange'].head()
```

```python
census1415['PerCapInc1415PctChange']=census1415['Per Capita Income_15']/census1415['Per Capita Income_14']-1
```

```python
census1415['PerCapInc1415PctChange'].head()
```

```python
census1516['PerCapInc1516PctChange']=census1516['Per Capita Income_16']/census1516['Per Capita Income_15']-1
```

```python
census1516['PerCapInc1516PctChange'].head()
```

```python
data_frames = [census1213, census1314, census1415, census1516]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Zipcode'],
                                            how='outer'), data_frames)
```

```python
df_merged.head()
```

```python
pctchanges=df_merged[["Zipcode","PerCapInc1213PctChange","PerCapInc1314PctChange", "PerCapInc1415PctChange", "PerCapInc1516PctChange"]]
```

```python
pctchanges.head()
```

```python
redata = pd.read_csv('censuspluszillowP2.csv')
```

```python
census12to17wPctChange=redata.merge(pctchanges, on="Zipcode")
```

```python
census12to17wPctChange.head()
```

```python
census12to17wPctChange.to_csv('census12to17wPctChange.csv')
```
